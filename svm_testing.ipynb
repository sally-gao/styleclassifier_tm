```python
import pandas as pd
import numpy as np
import pickle
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_fscore_support,accuracy_score,f1_score,precision_score,recall_score
from sklearn.model_selection import train_test_split
```

# Import Dataframe

```python
# Load pickled dataframe

#file = 'pickles/features_04_29.pkl'
file = 'pickles/features_04_30.pkl'

with open(file, 'rb') as f:
    df_features = pickle.load(f)
```

```python
df_features.head()
```

# SVM

```python
# X -> features, y -> label
X=df_features.loc[:, df_features.columns != 'source']

y=df_features.source
print(len(X))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y,test_size=0.3, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
clf = SVC(kernel = 'linear')

print("Fitting with default params ...")

clf.fit(X_train, y_train)

print("Getting test preds")
y_pred=clf.predict(X_test)

val_precision=precision_score(y_test,y_pred,average='macro')
val_recall=recall_score(y_test,y_pred,average='macro')
val_fscore=f1_score(y_test, y_pred, average='macro')
val_accuracy=accuracy_score(y_test, y_pred)

print("Getting train preds")
y_train_pred=clf.predict(X_train)
train_precision=precision_score(y_train,y_train_pred,average='macro')
train_recall=recall_score(y_train,y_train_pred,average='macro')
train_fscore=f1_score(y_train,y_train_pred, average='macro')
train_accuracy=accuracy_score(y_train,y_train_pred)
```

```python
print("Train Accuracy: ", train_accuracy, "\n",
     "Train Precision: ", train_precision, "\n",
     "Train Recall: ", train_recall, "\n",
     "Train F1 Score: ", train_fscore, "\n",
     "\n",
     "Test Accuracy: ", val_accuracy, "\n",
     "Test Precision: ", val_precision, "\n",
     "Test Recall: ", val_recall, "\n",
     "Test F1 Score: ", val_fscore)
```

# Varying C

```python
def fit_svm(c):
    
    print("Trying C= ", c, "\n")
    
    clf = SVC(kernel = 'linear', C = c)

    print("Fitting ...")

    clf.fit(X_train, y_train)

    print("Getting test preds")
    y_pred=clf.predict(X_test)

    val_precision=precision_score(y_test,y_pred,average='macro')
    val_recall=recall_score(y_test,y_pred,average='macro')
    val_fscore=f1_score(y_test, y_pred, average='macro')
    val_accuracy=accuracy_score(y_test, y_pred)

    print("Getting train preds\n")
    y_train_pred=clf.predict(X_train)
    train_precision=precision_score(y_train,y_train_pred,average='macro')
    train_recall=recall_score(y_train,y_train_pred,average='macro')
    train_fscore=f1_score(y_train,y_train_pred, average='macro')
    train_accuracy=accuracy_score(y_train,y_train_pred)
    
    print("Train Accuracy: ", train_accuracy, "\n",
     "Train Precision: ", train_precision, "\n",
     "Train Recall: ", train_recall, "\n",
     "Train F1 Score: ", train_fscore, "\n",
     "\n",
     "Test Accuracy: ", val_accuracy, "\n",
     "Test Precision: ", val_precision, "\n",
     "Test Recall: ", val_recall, "\n",
     "Test F1 Score: ", val_fscore, "\n\n")
```

```python
for c in [0.5, 1, 1.5]:
    fit_svm(c)
```

```python
for c in [0.01, 0.25, 3]:
    fit_svm(c)
```
